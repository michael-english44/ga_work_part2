```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Matplotlib was original plotting library
#Seaborn was built on matplotlib
#Seaborn has more advanced plotting techniques
##pyplot is high overview of matplotlib ex. does all formatting of bars, axis ticks
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.info()
#Interested in distributions of continuous variables
```

## univariate representations

## Strip plit in Seaborn

```python
sns.stripplot(x = tips['tip'], y= tips['day'])
```

```python
plt.hist(tips['tip'], bins = 20)
#Histogram is a range of values of countsof how many things are in each range
# Think count of things in each bin
# This is a discrete distribution
```

```python
#Smoothed version of histogram
#Kernel density estimation

sns.kdeplot(tips['tip'])
```

## Boxplot

```python
sns.boxplot(tips['tip'])
#seaborn identifys outliers
#outliers determined by
#Bars represent quartile boundaries excluding the outliers
# 5 number summary
```

```python
sns.boxplot(x = tips['tip'], y = tips['day'], hue = tips['sex'])
```

```python
sns.countplot(tips['sex'])
```

```python
#Is there a relationship between the total bill and the amount tipped?
plt.scatter(tips['total_bill'], tips['tip'], c = 'b')
```
